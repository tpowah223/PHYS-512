```python
import numpy as np, matplotlib.pyplot as plt
x=np.linspace(1e7,2e8, 100)
y_sub = np.sqrt(1+x**2)-x
y_nosub = 1/(np.sqrt(1+x**2)+x)

plt.plot(x, y_sub, label="subtracted form")
plt.plot(x, y_nosub, label="non-subtracted form")
z = abs(y_sub - y_nosub)
plt.plot(x, z, label="difference")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
