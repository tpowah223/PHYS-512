```python
import numpy as np , matplotlib.pyplot as plt
deltaX=10**np.linspace(-12,2, 200)
x = 1
eps = 1e-16

def f(x):
    return x**2

err1 =np.zeros_like(deltaX)
err2 =np.zeros_like(deltaX)

for i, dx in enumerate(deltaX):
    #first order forward difference
    dfdx = (f(x+dx)-f(x))/dx
    #df2dx = (2/dx**2)*(f(x+dx)-f(x)-dx*dfdx)
    err1[i] = np.abs((dfdx-2*x)/(2*x))
    #second order center difference
    dfdx_2 = (f(x+dx) - f(x-dx))/ (2*dx)
    #df2dx_2 = (dx**2)*(f(x+dx)-2*f(x)+f(x-dx))
    err2[i] = np.abs((dfdx_2 - 2*x)/(2*x))


plt.plot(deltaX,err1, label="first order fwd")
plt.plot(deltaX,err2, label="second order center")
plt.plot(deltaX, (eps*f(x))/deltaX, ":",label=r'$\epsilon f(x)/\Delta x$')
plt.plot(deltaX,deltaX, ":")
plt.plot(deltaX,deltaX**2, ":")
plt.title(r'$f(x) = x^2, \ \ \ x=%lg$' % (x,))
plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Fractional error in derivative')
plt.legend()
plt.show()
```
